```python
epsilon = 1e-6
```

```python
# Проведем некоторые преобразования для посл. обработки

# log1p для избежания log(0)
for col in ['triglyceride', 'Gtp', 'AST', 'ALT', 'fasting blood sugar', 'Cholesterol']:
    df[f'{col}_log'] = np.log1p(df[col] + epsilon)

# Возможно нелинейные признаки
df['age_sq'] = df['age'] ** 2
df['BMI'] = df['weight(kg)'] / (df['height(cm)'] / 100) ** 2 # Индекс массы тела
df['Weist_to_Height'] = df['waist(cm)'] / (df['height(cm)'] + epsilon)

# Медицинские соотношения (спасибо гугл)
# Коэф. де Ритиса (AST/ALT) - маркер поражения печени
df['AST_to_ALT'] = df['AST'] / (df['ALT'] + epsilon)
# Соотногение LDL/HDL - показатель сердечно-сосудистого риска
df['LDL_to_HDL'] = df['LDL'] / (df['HDL'] + epsilon)
# Соотношение триглицериды/HDL - индикатор инсулинорезистентности
df['TG_to_HDL'] = df['triglyceride'] / (df['HDL'] + epsilon)

# Взаимодействия
# У курильщиков ГГТ часто повышена
df['age_x_Gtp'] = df['age'] * df['Gtp']
# Курения усугубляет холестерин и давление
df['systolic_x_LDL'] = df['systolic'] * df['LDL']
# Метаболический риск (ожирение + курение)
df['HighRisk_Metabolic'] = ((df['BMI'] > 30) & (df['LDL_to_HDL'] > 3)).astype(int)

# Доп. бинарные признаки
# Индикатор повышенного ГГТ (алгоколь/курение)
df['Hight_Gtp'] = (df['Gtp'] > 50).astype(int)
# Индикатор низкого HDL
df['Low_HDL'] = (df['HDL'] > 50).astype(int)
# Индиктор плохого зрения (курение влияет на сосуды сетчатки)
df['Poor_Eyesight'] = ((df['eyesight(left)'] > 1.5) | (df['eyesight(right)'] > 1.5)).astype(int)
```

```python
# bool столбцы уже подходят, поэтому их мы вынесем
target = df['smoking']
number_features = df[df.select_dtypes(include=['number']).columns.drop('smoking')]
bools = df[df.select_dtypes(include=['bool']).columns]

# ------------------------------------------------------------------------

number_features_test = test[test.select_dtypes(include=['number']).columns]
bools_test = test[test.select_dtypes(include=['bool']).columns]
```

### MIN-MAX

```python
# x = (x - x_min) / (x_max-x_min), где x - признак
number_features_minmax_normalized = number_features.apply(lambda feature: (feature - feature.min()) / (feature.max() - feature.min()))

print(number_features_minmax_normalized.agg(['max', 'min']))

df_minmax_normalized = number_features_minmax_normalized.join(bools).join(target)

# ------------------------------------------------------------------------

number_features_minmax_normalized_test = number_features_test.apply(lambda feature: (feature - feature.min()) / (feature.max() - feature.min()))
test_minmax_normalized = number_features_minmax_normalized_test.join(bools_test)
```

### Z-SCORE

```python
# z = (x - 'среднее значение') / 'стандартное отклонение'

number_features_zscore_normalized = number_features.apply(lambda feature: (feature - feature.mean()) / feature.std())

print(number_features_zscore_normalized.agg(['std', 'mean']))

df_zscore_normalized = number_features_zscore_normalized.join(bools).join(target)

# ------------------------------------------------------------------------
number_features_zscore_normalized_test = number_features_test.apply(lambda feature: (feature - feature.mean()) / feature.std())
test_zscore_normalized = number_features_zscore_normalized_test.join(bools_test)
```

## Бэггинг

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
from sklearn.base import clone
from sklearn.metrics import accuracy_score
from sklearn.ensemble import BaggingClassifier as SklearnBaggingClassifier
from sklearn.utils import resample
```

```python
class BaggingClassifier:
    def __init__(self, base_estimator=None, n_estimators=10, random_state=None):
        self.base_estimator = base_estimator if base_estimator is not None else DecisionTreeClassifier()
        self.n_estimators = n_estimators
        self.random_state = random_state
        self.estimators_ = []

    def fit(self, X, y):
        np.random.seed(self.random_state)
        self.estimators_ = []

        for _ in range(self.n_estimators):
            X_samples, y_sample = resample(X, y, random_state=np.random.randint(1e5))

            model = clone(self.base_estimator)
            model.fit(X_samples, y_sample)
            self.estimators_.append(model)

    def predict(self, X):
        predictions = np.array([model.predict(X) for model in self.estimators_])

        return np.apply_along_axis(
            lambda x: np.bincount(x).argmax(),
            axis=0,
            arr=predictions.astype(int)
        )
```

```python
tree = DecisionTreeClassifier(max_depth=3)

my_bagging = BaggingClassifier(tree, n_estimators=50, random_state=RANDOM_STATE)
my_bagging.fit(X_train, y_train)
my_pred = my_bagging.predict(X_test)
my_accuracy = accuracy_score(y_test, my_pred)

sklearn_bagging = SklearnBaggingClassifier(tree, n_estimators=50, random_state=RANDOM_STATE, bootstrap=True)
sklearn_bagging.fit(X_train, y_train)
sklearn_pred = sklearn_bagging.predict(X_test)
sklearn_accuracy = accuracy_score(y_test, sklearn_pred)

print(f'Моя точность:{my_accuracy:.3f}')
print(f'Sklearn точность:{sklearn_accuracy:.3f}')
```

## Градиентый бустинг

```python
from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score
from scipy.special import expit
```

```python
class GradientBoostingClassifier:
    def __init__(self, n_estimators=100, learning_rate=0.1, max_depth=3, random_state=None):
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.random_state = random_state
        self.max_depth = max_depth
        self.estimators_ = []
        self.initial_pred = None

    def fit(self, X, y):
        np.random.seed(self.random_state)

        pos_class_prob = np.mean(y)
        self.initial_pred = np.log(pos_class_prob / (1 - pos_class_prob))
        predictions = np.full_like(y, self.initial_pred, dtype=np.float64)

        for _ in range(self.n_estimators):
            probas = expit(predictions)
            residuals = y - probas

            tree = DecisionTreeRegressor(max_depth=self.max_depth, random_state=RANDOM_STATE)
            tree.fit(X, residuals)

            predictions += self.learning_rate * tree.predict(X)

            self.estimators_.append(tree)

        return self

    def predict_proba(self, X):
        predictions = np.full(X.shape[0], self.initial_pred, dtype=np.float64)

        for tree in self.estimators_:
            predictions += self.learning_rate * tree.predict(X)

        probas = expit(predictions)
        return np.vstack([1 - probas, probas]).T

    def predict(self, X):
        probas = self.predict_proba(X)
        return (probas[:, 1] > 0.5).astype(int)
        
```

```python
from sklearn.ensemble import GradientBoostingClassifier as SklearnGradientBoostingClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
import time
```

```python
params = {"n_estimators": 100, 'learning_rate': 0.1, 'max_depth': 3, 'random_state': RANDOM_STATE}

models = {
    'My Gb': GradientBoostingClassifier(**params),
    'Sklearn GB': SklearnGradientBoostingClassifier(**params),
    'XGBoost': XGBClassifier(**params),
    'LightGBM': LGBMClassifier(**params),
    'CatBoost': CatBoostClassifier(**params)
}

results = []

for name, model in models.items():
    start_time = time.time()
    model.fit(X_train, y_train)
    train_time = time.time() - start_time

    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]

    results.append({
        'Model': name,
        'Accuracy': accuracy_score(y_test, y_pred),
        'F1 score': f1_score(y_test, y_pred),
        'ROC AUC': roc_auc_score(y_test, y_proba),
        'Train Time (ms)': train_time
    })

results_df = pd.DataFrame(results).sort_values('ROC AUC', ascending=False)
print(results_df.to_markdown(index=False))
```

## Optuna

```python
import optuna
from sklearn.model_selection import cross_val_score
from xgboost import XGBClassifier

def objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 500),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.001, 1, log=True),
        'max_depth': trial.suggest_int('max_depth', 3, 12),
        'num_leaves': trial.suggest_int('num_leaves', 20, 200)
        'subsamples': trial.suggest_float('subsamples', 0.6, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0),
        'gamma': trial.suggest_float('gamma', 0, 1),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 10),
        'min_child_samples': trial.suggest_int('min_child_weight', 5, 100),
        'random_state': RANDOM_STATE,
        'n_jobs': -1
    }

    model = XGBClassifier(**params)
    score = cross_val_score(model, X_train, y_train, cv=5, scoring='roc_auc').mean()
    return score
```

```python
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=30, timeout=300)

trial = study.best_trial

print("Best trial:")
print(f"AUC ROC: {trial.value:.5f}")
print(" Params: ")
for key, value in trial.params.items():
    print(f"  {key}: {value}")
```
